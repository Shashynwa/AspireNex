```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Dataset

```python
df = pd.read_csv('creditcard.csv')
df
```

```python
df.info()
```

### Missing Values

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
print('No Frauds', round(df['Class'].value_counts()[0]/len(df) * 100,2), '% of the dataset')
print('Frauds', round(df['Class'].value_counts()[1]/len(df) * 100,2), '% of the dataset')
```

### Taking Equal amt of data

```python
df = df.sample(frac=1)

fraud_df = df.loc[df['Class'] == 1]
non_fraud_df = df.loc[df['Class'] == 0][:492]

normal_distributed_df = pd.concat([fraud_df, non_fraud_df])

new_df = normal_distributed_df.sample(frac=1, random_state=42)

new_df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, roc_auc_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

X = new_df.drop('Class', axis=1)
y = new_df['Class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

### Logistic Regression

```python
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
```

```python
y_pred = log_reg.predict(X_test)

print('Confusion Matrix: \n', confusion_matrix(y_test, y_pred))
print('Accuracy: ', accuracy_score(y_test, y_pred))
print('Classification Report: \n', classification_report(y_test, y_pred))
print('ROC AUC Score: ', roc_auc_score(y_test, y_pred))
```

### Decision Tree

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
y_pred = dt.predict(X_test)

print('Confusion Matrix: \n', confusion_matrix(y_test, y_pred))
print('Accuracy: ', accuracy_score(y_test, y_pred))
print('Classification Report: \n', classification_report(y_test, y_pred))
print('ROC AUC Score: ', roc_auc_score(y_test, y_pred))
```

```python
from sklearn import tree
plt.figure(figsize=(20,10))
tree.plot_tree(dt, filled=True)
```

### Random Forest

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)

print('Confusion Matrix: \n', confusion_matrix(y_test, y_pred))
print('Accuracy: ', accuracy_score(y_test, y_pred))
print('Classification Report: \n', classification_report(y_test, y_pred))
print('ROC AUC Score: ', roc_auc_score(y_test, y_pred))
```

### Oversampling using SMOTE

```python
from imblearn.over_sampling import SMOTE

X = df.drop('Class', axis=1)
y = df['Class']

smote = SMOTE(sampling_strategy='minority')
X_sm, y_sm = smote.fit_resample(X, y)

print('No Frauds', round(y_sm.value_counts()[0]/len(y_sm) * 100,2), '% of the dataset')
print('Frauds', round(y_sm.value_counts()[1]/len(y_sm) * 100,2), '% of the dataset')
```

### Logistic Regression,Decision Tree and Random Forest Models

```python
lr = LogisticRegression()
dt = DecisionTreeClassifier()
rf = RandomForestClassifier(n_estimators=50,max_depth=10,min_samples_leaf=10,min_samples_split=50,n_jobs=-1,random_state=42)

X_train, X_test, y_train, y_test = train_test_split(X_sm, y_sm, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)

print('Confusion Matrix: \n', confusion_matrix(y_test, y_pred))
print('Accuracy: ', accuracy_score(y_test, y_pred))
print('Classification Report: \n', classification_report(y_test, y_pred))
print('ROC AUC Score: ', roc_auc_score(y_test, y_pred))
```

```python
dt.fit(X_train, y_train)
```

```python
y_pred = dt.predict(X_test)

print('Confusion Matrix: \n', confusion_matrix(y_test, y_pred))
print('Accuracy: ', accuracy_score(y_test, y_pred))
print('Classification Report: \n', classification_report(y_test, y_pred))
print('ROC AUC Score: ', roc_auc_score(y_test, y_pred))
```

```python
y_pred = rf.predict(X_test)

print('Confusion Matrix: \n', confusion_matrix(y_test, y_pred))
print('Accuracy: ', accuracy_score(y_test, y_pred))
print('Classification Report: \n', classification_report(y_test, y_pred))
print('ROC AUC Score: ', roc_auc_score(y_test, y_pred))
```
